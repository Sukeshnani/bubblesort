Part 1. 
Use Python to write a “Bubble Sort” function to sort the text from chapter 1 of Alice in Wonderland by 
Lewis Carroll.   This text can be found at https://www.gutenberg.org/files/11/11-0.txt    In this textfile,  
chapter 1 starts on line 56 and ends on line 270 – there are 2415 words in this chapter – note that the 
“astericks” are being counted as words.     Use your Bubble Sort function to sort the total text in 
chapter 1,  then print this sorted text.  IMPORTANT – write your own sort function – do not use 
Python’s “built-in” sort function.  

```python
import matplotlib.pyplot as pp
import urllib.request
import time as t
import pandas as pd

l = 0
tx = ""
for lines in urllib.request.urlopen("https://www.gutenberg.org/files/11/11-0.txt"):
    l= l+1
    if l>=56 and l<=270:
        tx+= lines.decode('utf-8').replace('\r\n',' ')
        ls = list(tx.split(" "))
while('' in ls):
 ls.remove('')
print(len(ls))
```

```python
#PART1
def bubblesort(v1):
    for n in range(len(v1)):
        swap = False
        v = 0
        while v < len(v1)-1:
            if v1[v]>v1[v+1]:
                v1[v],v1[v+1] = v1[v+1],v1[v]
                swap = True
            v = v+1
        if swap == False:
            break
    return(v1)   
 
```

```python
bubblesort(ls)
```

#Part 2.   
Use Python’s time module to time how long your function takes to sort all the total text in chapter 1.   
Print out this timing value. 

```python
#PART2
import time as t
def duration(v1):
    st = t.perf_counter()
    #time after sorting text
    t1 = bubblesort(v1)
    et = t.perf_counter()
    return et-st
print('''"function takes to sort all the total text in chapter 1 is"''',duration(ls))
```

#Part 3.   
Chapter 1 has 2415 words in the chapter.   Create an X-Y graph that plots how long your Bubble Sort 
takes to sort 5 text segments,  where the size of the text segments increment 483 words per segment.   
Your Y-axis would indicate time in seconds;  Your X-axis will have values that represent 5 text segments,  
where the words in the text segments increment by amounts of 483, 966, 1499, 1932, 2415 words in 
each of the 5 text segments.    This graph would consist of 5 (X, Y) pairs of values -  (seg_1_time,  
num_words_seg_1), ..... , (seg_5_time, num_words_seg_5).    Describe the shape of this line plot – does 
it increase in a straight line fashion, or is the line curvilinear?

```python
#PART3
text_segments = int(len(ls)/5)
#time calculation for 5 segments with values
tm = []
seg = []
for i in range(5):
  segs = ls[0:text_segments*(i+1)]
  tm.append(duration(segs))
  seg.append(text_segments*(i+1))
print(tm,seg)
```

```python
from pandas.core.frame import DataFrame
plt = {'s':seg, 'v':tm }
df = pd.DataFrame(plt,columns=['s','v'])
pp.plot(df['s'],df['v'], color = "green", marker = "o")
pp.title("Segments and Duration", fontsize=12)
pp.xlabel('segments',fontsize=12)
pp.ylabel('duration',fontsize=12)
pp.xticks(seg,seg)
pp.grid(True)
print("The graph is in is the line curvilinear.")
pp.show()
```
